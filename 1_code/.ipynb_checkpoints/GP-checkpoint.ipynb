```python
import pandas as pd
import numpy as np
import matplotlib.pylab as plt
from sklearn.model_selection import train_test_split
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import RBF, ConstantKernel,WhiteKernel,RationalQuadratic, Matern, DotProduct
```

```python
ds_demo = pd.read_csv("twitter_income.csv")
y = np.log(ds_demo.loc[:,"mean_income"])
X = ds_demo.loc[:,['Topic_163','Topic_139','followers','Topic_124',
 'Topic_95','Topic_162','listed','Topic_196','Topic_173','foll_friend_ratio','enlish_tweets',
 'Topic_48','age','Topic_32','Topic_36','Topic_28','Topic_34','Topic_56','Topic_105','Topic_144']]
X_train,X_test,y_train,y_test=train_test_split(X, y, test_size=0.2, random_state=999)
```

```python
kernel = 1.* RBF(length_scale=1., length_scale_bounds=(1e-6, 1e6)) \
    + WhiteKernel(noise_level=1., noise_level_bounds=(1e-10, 1e20)) \
    + RationalQuadratic(length_scale=1.) \
    
gp = GaussianProcessRegressor(kernel=kernel, random_state=999)

gp.fit(X_train, y_train)

y_pred, sigma = gp.predict(X_test, return_std=True)

print(gp.log_marginal_likelihood_value_)
print(gp.kernel_)
```

```python
plt.plot(y_pred, y_test, '.')
```

```python
plt.hist(y_pred, bins = 10, alpha = 0.5);
plt.hist(y_test, bins = 10, alpha = 0.5);
```
